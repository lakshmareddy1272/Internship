```python
# 2) Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame

imdburl = 'https://www.imdb.com/chart/top/'
source = requests.get(imdburl)
source.raise_for_status()           # raise_for_status() will check for url availability

soup = BeautifulSoup(source.text, 'html.parser')

moviename = []
releaseyear = []
movierating = []

movies = soup.find('tbody', class_="lister-list").find_all('tr')
for movie in movies:
    name = movie.find('td', class_="titleColumn").a.text
    moviename.append(name)
    year = movie.find('td', class_="titleColumn").span.text.strip('()')
    releaseyear.append(year)
    rating = movie.find('td', class_="ratingColumn imdbRating").strong.text
    movierating.append(rating)

movie_df=pd.DataFrame({'Movie_Name':moviename[:100],'Rating':movierating[:100],'Year_Of_Release':releaseyear[:100]})
movie_df.index=range(1,len(movie_df)+1)
movie_df
```

```python
# 4) Write a python program to scrape product name, price and discounts from https://meesho.com/bags-ladies/pl/p7vbp .

url = 'https://meesho.com/bags-ladies/pl/p7vbp'
ms_source = requests.get(url)
ms_source

soup=BeautifulSoup(ms_source.content,'html.parser')

productname = []
price = []
discount = []

ms_data = soup.find_all('div', class_="Card__BaseCard-sc-b3n78k-0 fa-dWHP NewProductCard__CardStyled-sc-j0e7tu-0 fqpiyA NewProductCard__CardStyled-sc-j0e7tu-0 fqpiyA")
for lbag in ms_data:
    productname.append(lbag.find('p',class_="Text__StyledText-sc-oo0kvp-0 bWSOET NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 cQhePS NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 cQhePS").text.strip())
    price.append(lbag.find('h5', class_="Text__StyledText-sc-oo0kvp-0 hiHdyy").text)
    discount.append(lbag.find('p', class_="Text__StyledText-sc-oo0kvp-0 fCJVtz NewProductCard__DiscountTextParagraph-sc-j0e7tu-16 kmYsnm NewProductCard__DiscountTextParagraph-sc-j0e7tu-16 kmYsnm").text)

meesho_df=pd.DataFrame({'Product_Name':productname, 'Price':price, 'Discount':discount})
meesho_df.index=range(1,len(meesho_df)+1)
meesho_df
```

```python
#5 Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape:

## 5a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

odim_source = requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi','https://www.icc-cricket.com/rankings/womens/team-rankings/odi')
soup = BeautifulSoup(odim_source.text,'html.parser')

team = []
matches = []
points = []
rating = []

parent=soup.find('tbody')
content=parent.find_all('tr')

for omtr in content:  # omtr - odi men team rank
    team.append(omtr.find_all('td')[1].find_all('span')[1].text)
    matches.append(omtr.find_all('td')[2].text)
    points.append(omtr.find_all('td')[3].text)
    rating.append(omtr.find_all('td')[4].text.strip())

ODI_Men_Team_Ranking_dict = {'Team':team[:10], 'Matches':matches[:10], 'Points':points[:10], 'Rating':rating[:10]}
ODI_Men_Team_Ranking_df=pd.DataFrame(ODI_Men_Team_Ranking_dict)
ODI_Men_Team_Ranking_df.index = range(1,len(ODI_Men_Team_Ranking_df)+1)
ODI_Men_Team_Ranking_df
```

```python
# 7) Write a python program to scrape details of all the posts from coreyms.com Scrape the heading, date, content and the code for the video from the link for the youtube video from the post.

coreyms_url = requests.get('https://coreyms.com/')
coreyms_url

soup=BeautifulSoup(coreyms_url.content,'html.parser')

heading = []
date = []
content = []
videolink = []

coreyms_data = soup.find_all('article')
for data in coreyms_data:
    heading.append(data.find('a', class_="entry-title-link").text.strip())
    date.append(data.find('time', class_="entry-time").text.replace(',',''))
    content.append(data.find('div', class_="entry-content").p.text)
    
    if data.find('iframe'):
        video_link=data.find('iframe',class_="youtube-player")['src']
        v_lnk=f"https://youtube.com/watch?v={video_link.split('/')[4].split('?')[0]}"
    else:
        v_lnk = 'Video Link Not Available'
    videolink.append(v_lnk)
coreyms_df=pd.DataFrame({'Heading':heading,'Date':date,'Content':content, 'VideoLink':videolink})
coreyms_df.index=range(1,len(coreyms_df)+1)
coreyms_df
```

```python
# 9) Write a python program to scrape mentioned details from dineout.co.in
       # i) Restaurant name
       # ii) Cuisine
       # iii) Location
       # iv) Ratings
       # v) Image URL
    

dineouturl = requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
dineouturl

soup=BeautifulSoup(dineouturl.content, 'html.parser')

Restaurant_name = []
Cuisine = []
Location = []
Rating = []
Imageurl = []

for name in soup.find_all('div', class_="restnt-info cursor"):
    Restaurant_name.append(name.a.text)
for cui in soup.find_all('span', class_="double-line-ellipsis"):
    Cuisine.append(cui.a.text)
for loc in soup.find_all('div', class_="restnt-loc ellipsis"):
    Location.append(loc.text)
for rat in soup.find_all('div', class_="restnt-rating rating-4"):
    Rating.append(rat.text)
for img in soup.find_all('img', class_="no-img"):
    Imageurl.append(img['data-src'])
    
dineout_df=pd.DataFrame({'Restaurant_Name':Restaurant_name,
                         'Cuisine':Cuisine,
                         'Location':Location,
                         'Ratings':Rating,
                         'Image_URL':Imageurl})

dineout_df.index=range(1,len(dineout_df)+1)
dineout_df
```
